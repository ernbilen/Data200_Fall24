```python
from selenium import webdriver
import pandas as pd
import random
import time
pd.set_option('display.max_colwidth', None)

# Task 1: Best Books analysis
# This code grabs book data from the best books page
page_data = pd.DataFrame()
for i in range(1,13):
    driver = webdriver.Chrome()
    driver.get(f'https://www.goodreads.com/list/best_of_year/2023?id=183940.Best_Books_of_2023&page={i}')
    time.sleep(random.uniform(.1,.2))
    try:
        popup = driver.find_element('xpath', '/html/body/div[3]/div/div/div[1]/button/img')
        time.sleep(random.uniform(.1,.2))
        popup.click()
        print("Popup closed.")
    except:
        pass
    
    table_element = driver.find_elements('xpath', "//tr[contains(@itemtype, 'schema.org/Book')]")
    time.sleep(random.uniform(.2,.4))

    table = []
    for i in table_element:
        lines = i.text.split('\n')
        # Filter out any unwanted lines (e.g., "really liked it")
        lines = [line for line in lines if line.strip().lower() != 'really liked it']
        # Now take the first 5 lines of the cleaned list
        row_data = lines[:5]

        # Extract the author's name (between 'by' and '(')
        author_full = row_data[2]  # e.g., 'by Rebecca Yarros (Goodreads Author)'
        author_name = author_full.split('by ')[-1].split(' (')[0].strip()  # Extract author name
        row_data[2] = author_name  # Replace the full 'Author' field with just the name

        # Split the Rating field into Average and Total Ratings
        rating_text = row_data[3]  # e.g., '4.57 avg rating ‚Äî 2,291,054 ratings'
        avg_rating = rating_text.split(' avg rating')[0]  # Extract '4.57'
        total_ratings = rating_text.split('‚Äî')[-1].strip().split(' ratings')[0]  # Extract '2,291,054'
        total_ratings = total_ratings.replace(',', '')  # Remove commas for numerical processing
        row_data[3] = avg_rating  # Replace Rating with Average Rating
        row_data.append(total_ratings)  # Add Total Ratings as a new field

        # Split the Score field into Score and Total Votes
        score_text = row_data[4]  # e.g., 'score: 53,258, and 536 people voted'
        score_value = score_text.split('score: ')[-1].split(', and')[0].strip()  # Extract '53,258'
        total_votes_text = score_text.split('and ')[-1]  # e.g., '536 people voted' or '1 person voted'
        total_votes = total_votes_text.replace(' person voted', '').replace(' people voted', '').strip()  # Remove text
        total_votes = total_votes.replace(',', '')  # Remove commas for numerical processing
    
        score_value = score_value.replace(',', '')  # Remove commas for numerical processing
        row_data[4] = score_value  # Replace Score with the cleaned Score value
        row_data.append(total_votes)  # Add Total Votes as a new field
        
        href = i.find_element('xpath', './/td[3]/a').get_attribute('href')
        row_data.append(href)  # Add href to the row data

        table.append(row_data)

    # Create the DataFrame with the updated columns
    df = pd.DataFrame(table, columns=['Rank', 'Title', 'Author', \
                                      'Average Rating', 'Score', 'Total Ratings', 'Total Votes','url'])
    page_data=pd.concat([df,page_data],axis=0)
    
    driver.quit()
page_data['Rank'] = page_data.Rank.astype('int')
page_data = page_data.sort_values('Rank').reset_index(drop=True)
```

```python
# This code continues the previous work and grabs more detailed book data by going to each book's url
import time
import random
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options

# Optional: Run in headless mode
chrome_options = Options()
chrome_options.add_argument("--headless")

books_data = pd.DataFrame()
script_start_time = time.time()  # Record the script's start time

for i in range(len(page_data)):
    driver = webdriver.Chrome(options=chrome_options)
    driver.get(page_data.url[i])
    start_time = time.time()  # Start the timer for this page
    timeout = 5  # Timeout in seconds for the entire page processing

    while True:
        try:
            time.sleep(random.uniform(0.2, 0.4))

            # Check and close the popup if it exists
            try:
                close_button = driver.find_element('xpath', '//button[@aria-label="Close"]')
                if close_button.is_displayed():
                    close_button.click()
            except Exception:
                pass  # Continue if the popup doesn't exist

            # Attempt to expand book details
            try:
                button = driver.find_element('xpath', '//button[@aria-label="Book details and editions"]')
                driver.execute_script("arguments[0].scrollIntoView(true);", button)
                time.sleep(random.uniform(0.2, 0.4))  # Randomized wait for scrolling
                button.click()
            except Exception:
                # If expanding book details fails, default to empty values and continue
                language = ''
                num_pages = ''
                break  # Exit the loop since expanding details failed

            time.sleep(random.uniform(0.2, 0.4))

            # Extract book details
            try:
                edition_details = driver.find_element('xpath', '//*[@class="EditionDetails"]').text
                language = edition_details.split("Language")[-1].split('\n')[1].strip() if "Language" in edition_details else ''
                num_pages = int(edition_details.split(' pages')[0].split('\n')[-1])
            except:
                language = ''
                num_pages = ''

            title = driver.find_element('xpath', '//*[@data-testid="bookTitle"]').text
            publication_date = driver.find_element('xpath', '//*[@data-testid="publicationInfo"]').text
            publication_date = publication_date.replace('First published', '').strip()
            author = driver.find_element('xpath', '//*[@data-testid="name"]').text
            genre = driver.find_element('xpath', '//*[@aria-label="Top genres for this book"]').text.split()[1]
            avg_rating = float(driver.find_element('xpath', '//*[@class="RatingStatistics__rating"]').text)
            num_ratings = driver.find_element('xpath', '//*[@class="RatingStatistics__meta"]').text
            num_ratings = int(num_ratings.split('ratings')[0].strip().replace(',', ''))

            # Handle missing currentlyReadingSignal
            try:
                currently_reading = driver.find_element('xpath', '//*[@data-testid="currentlyReadingSignal"]').text
                currently_reading = currently_reading.split(' people')[0].replace('m', '000000').replace('k', '000').replace('.', '').replace(',', '')
            except:
                currently_reading = ''

            # Handle missing toReadSignal
            try:
                wants = driver.find_element('xpath', '//*[@data-testid="toReadSignal"]').text
                wants = wants.split(' people')[0].replace('m', '000000').replace('k', '000').replace('.', '').replace(',', '')
            except:
                wants = ''

            # Add book data to DataFrame
            book_data = {
                'Title': title,
                'Author': author,
                'Average Rating': avg_rating,
                'Total Ratings': num_ratings,
                'Language': language,
                'Number of Pages': num_pages,
                'Publication Date': publication_date,
                'Genre': genre,
                'Currently Reading': currently_reading,
                'Wants to Read': wants,
                'URL': page_data.url[i]
            }

            books_data = pd.concat([books_data, pd.DataFrame([book_data])], ignore_index=True)

            # Save to CSV dynamically after each scrape
            books_data.to_csv("books_data.csv", index=False)

            # Calculate elapsed time since script started
            elapsed_time_minutes = (time.time() - script_start_time) / 60
            # Print the nth book scraped with title, author, and elapsed time
            print(f"{len(books_data)}th book scraped: '{title}' by {author} | Elapsed time: {elapsed_time_minutes:.2f} minutes")

            break  # Exit the loop if successful

        except Exception as e:
            # Check if timeout is exceeded
            elapsed_time = time.time() - start_time
            if elapsed_time > timeout:
                driver.refresh()
                start_time = time.time()  # Reset the timer after refresh
            else:
                pass  # Suppress error messages and retry

    driver.quit()

# Reset the index of the final DataFrame
books_data.reset_index(drop=True, inplace=True)
```

```python
# Age and Average rating

# Plot for all books
to_plot_all = books_data[(books_data.age > 26) & (books_data.age < 78)]
to_plot_all.loc[to_plot_all['Average Rating'] == 0, 'Average Rating'] = np.nan
to_plot_all = to_plot_all.groupby('age').agg({'Average Rating': 'mean', 'age': 'count'})\
.rename(columns={'age': 'count'}).reset_index()

# Plot for English books
to_plot_english = books_data[(books_data.age > 26) & (books_data.age < 78)]
to_plot_english = to_plot_english[to_plot_english.Language == 'English']
to_plot_english.loc[to_plot_english['Average Rating'] == 0, 'Average Rating'] = np.nan
to_plot_english = to_plot_english.groupby('age').agg({'Average Rating': 'mean', 'age': 'count'})\
.rename(columns={'age': 'count'}).reset_index()

# Plot for translated books
to_plot_translated = books_data[(books_data.age > 26) & (books_data.age < 78)]
to_plot_translated = to_plot_translated[(to_plot_translated.Language != 'English') & \
                                        (to_plot_translated.Language.notna())]
to_plot_translated.loc[to_plot_translated['Average Rating'] == 0, 'Average Rating'] = np.nan
to_plot_translated = to_plot_translated.groupby('age').agg({'Average Rating': 'mean', 'age': 'count'})\
.rename(columns={'age': 'count'}).reset_index()

# Optional: Plot all three on the same graph
plt.plot(to_plot_all['age'], to_plot_all['Average Rating'], label='All Books',color='black')
plt.plot(to_plot_english['age'], to_plot_english['Average Rating'], label='English Books')
plt.plot(to_plot_translated['age'], to_plot_translated['Average Rating'], label='Translated Books')
plt.xlim(25, 80)
plt.ylim(3, 5)
plt.xlabel("Stephen King's age")
plt.ylabel('Avg Rating')
plt.legend()
plt.show()
```
