```python
def getmovie(year):
    url = "https://api.themoviedb.org/3/discover/movie/?api_key="+apiKey+\
"&primary_release_year={}&sort_by=revenue.desc".format(year)
    out = requests.get(url).json()
    out = pd.DataFrame.from_dict(out['results'])
    out['year'] = year
    return out
```

```python
df=[]
for i in range(1960,1990):
    df.append(getmovie(i))

df_out = pd.concat(df)
df_out
```

```python
# check average vote by year
df_out.groupby('year').mean()[['vote_average','vote_count']]
```
