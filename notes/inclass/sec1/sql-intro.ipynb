```python
%%sql
-- How to return land size of each continent in a table?
SELECT region, SUM(land) AS total_land FROM countries
GROUP BY region
ORDER BY total_land DESC
```

```python
%%sql
-- How to return land size of each continent in a table?
-- In addition, you can place further filtering after group by
-- using HAVING
SELECT region, SUM(land) AS total_land FROM countries
GROUP BY region
HAVING total_land > 20000000
ORDER BY total_land DESC
```

```python
%%sql
-- How to return land size of each continent in a table?
-- In addition, you can place further filtering after group by
-- using HAVING
SELECT region, SUM(land) AS total_land FROM countries
WHERE land > 1000000
GROUP BY region
```

```python
%%sql
-- Can you return only the largest region
SELECT region, SUM(land) AS total_land FROM countries
GROUP BY region
HAVING total_land > 20000000
ORDER BY total_land DESC
LIMIT 1
```

```python
%%sql
-- return the instructorid, first and last name, dept id, and department's divison
-- for the insructors who taught the classes with the following classid: 21014, 21088, 21256, 21444
SELECT instructorid,instructorfirst,instructorlast,departmentid,division FROM departments
INNER JOIN instructors
USING(departmentid)
INNER JOIN instructor_class
USING(instructorid)
WHERE classid IN ('21014','21088','21256','21444')
```

```python
%%sql
--equivalently:
SELECT instructors.instructorid, instructors.instructorfirst, instructors.instructorlast,
departments.departmentid, departments.departmentname FROM departments
INNER JOIN instructors
ON departments.departmentid = instructors.departmentid
INNER JOIN instructor_class
ON instructors.instructorid = instructor_class.instructorid
WHERE classid=21014 OR classid=21088 OR classid=21256 OR classid=21444
```

```python
%%sql
-- return a table that shows these same courses in the following format:
-- column1: 4lettercoursesubject-coursenumber-section
-- column2: class meeting times for the same classes: '21014','21088','21256','21444
-- Note: use "||" which is "+"  in SQL
SELECT coursesubject || "-" || coursenum || "-" || classsection AS classinfo, classmeeting FROM classes
LIMIT 5
```

```python
%%sql
-- extension to the table above: return a third column "enrollment" showing how many students each section has
-- Note: count the students from the student_class table using COUNT(); use GROUP BY
SELECT studentid,classid, coursesubject || "-" || coursenum || "-" || classsection AS classinfo, classmeeting, COUNT(studentid) FROM classes
INNER JOIN student_class
USING(classid)
GROUP BY classid
LIMIT 10
```

```python
%%sql
-- find out if there are countries whose land observations exist but import or export obs don't exist
SELECT * FROM countries
LEFT JOIN 
indicators
USING(code)
GROUP BY code
HAVING (imports IS NULL OR exports IS NULL) AND (land IS NOT NULL)
```

```python
%%sql
-- find out if there are countries whose import or export obs exist but land observations don't exist 
SELECT * FROM indicators
LEFT JOIN 
countries
USING(code)
GROUP BY code
HAVING (land IS NULL) AND (imports IS NOT NULL OR exports is NOT NULL)
```
