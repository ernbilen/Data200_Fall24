```python
comment = driver.find_element('xpath','//*[@id="Comment_1726631"]/div/div[3]/div/div[1]').text
comment
```

```python
# get all comment ids from a single page
ids = driver.find_elements('xpath','//*[contains(@id,"Comment_")]')

comment_ids = []
for i in ids:
    comment_ids.append(i.get_attribute('id'))
```

```python
comment_ids
```

```python
comments = []
for j in comment_ids:
    print(j)
    username = driver.find_element('xpath','//*[@id="' + j + '"]/div/div[2]/div[1]/span[1]/a[2]').text
    date = driver.find_element('xpath','//*[@id="' + j + '"]/div/div[2]/div[2]/span/a/time').get_attribute('title')
    avatar = driver.find_element('xpath','//*[@id="'+j+'"]/div/div[2]/div[1]/span[1]/a[1]/img').get_attribute('src')
    comment_text = driver.find_element('xpath','//*[@id="'+j+'"]/div/div[3]/div/div[1]').text
    comments.append([username,date,avatar,comment_text])
```

```python
df = pd.DataFrame(comments,columns=['username','date','avatar','comment_text'])
df
```
