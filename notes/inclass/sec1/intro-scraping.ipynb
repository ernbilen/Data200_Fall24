```python
# multi page scrape
# first option: loop over urls using page numbering

import pandas as pd
import numpy as np
from selenium import webdriver
import time
import random

def get_comment_ids():
    ids = driver.find_elements('xpath','//*[contains(@id , "Comment_")]')
    comment_ids = []
    for i in ids:
        comment_ids.append(i.get_attribute('id'))
    return comment_ids

mylist = range(1,10)

df = pd.DataFrame() # declares an empty df to be populated
for k in mylist:
    driver = webdriver.Chrome()
    driver.get(f'https://forums.edmunds.com/discussion/2864/general/x/entry-level-luxury-performance-sedans/p{k}')
    time.sleep(random.uniform(.7,1.2)) # this does two things: 1) mimicks human behavior 2) waits for the website to return info before making xpath searches 
    comment_ids = get_comment_ids()
    
    comments = []
    for j in comment_ids:
        print(k,j)
        usernames = driver.find_element('xpath','//*[@id="'+ j + '"]/div/div[2]/div[1]/span[1]/a[2]').text
        date = driver.find_element('xpath','//*[@id="' + j + '"]/div/div[2]/div[2]/span/a/time').get_attribute('title')
        avatar=driver.find_element('xpath','//*[@id="'+ j + '"]/div/div[2]/div[1]/span[1]/a[1]/img').get_attribute('src')
        comment_text = driver.find_element('xpath','//*[@id="' + j + '"]/div/div[3]/div').text
        comments.append([usernames,date,avatar,comment_text])
        
    df_single = pd.DataFrame(comments,columns=['username','date','avatar','comment_text'])
    df = pd.concat([df_single,df]) # concatenates single page data to the main df
    
    #driver.close() # optional, in case you want to make sure that the driver is closed after scraping a single page

```

```python
# good practice: use try except blocks to catch errors

# multi page scrape
# option 2: have the driver press "next page" button

import pandas as pd
import numpy as np
from selenium import webdriver
import time
import random

def get_comment_ids():
    ids = driver.find_elements('xpath','//*[contains(@id , "Comment_")]')
    comment_ids = []
    for i in ids:
        comment_ids.append(i.get_attribute('id'))
    return comment_ids

driver = webdriver.Chrome()
driver.get(f'https://forums.edmunds.com/discussion/2864/general/x/entry-level-luxury-performance-sedans')
    
df = pd.DataFrame() # declares an empty df to be populated
for k in range(1,436):
    time.sleep(random.uniform(.7,1.2)) # this does two things: 1) mimicks human behavior 2) waits for the website to return info before making xpath searches 
    comment_ids = get_comment_ids()
    
    comments = []
    for j in comment_ids:
        print(k,j)
        try:
            usernames = driver.find_element('xpath','//*[@id="'+ j + '"]/div/div[2]/div[1]/span[1]/a[2]').text
        except:
            usernames = ''
        try:
            date = driver.find_element('xpath','//*[@id="' + j + '"]/div/div[2]/div[2]/span/a/time').get_attribute('title')
        except:
            date = ''
        try:
            avatar=driver.find_element('xpath','//*[@id="'+ j + '"]/div/div[2]/div[1]/span[1]/a[1]/img').get_attribute('src')
        except:
            avatar=''
        try:
            comment_text = driver.find_element('xpath','//*[@id="' + j + '"]/div/div[3]/div').text
        except:
            comment_text=''
        comments.append([usernames,date,avatar,comment_text])
        
    df_single = pd.DataFrame(comments,columns=['username','date','avatar','comment_text'])
    df = pd.concat([df_single,df]) # concatenates single page data to the main df
    
    #df.to_csv('comments.csv')
    
    button = driver.find_element('xpath','//*[contains(@title , "Next Page")]')
    #button = driver.find_element('xpath','//*[(@title , "Next Page")]') # alternative
    button.click()
    
    
```
